```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
import seaborn as sbn
```

### Loading dataset

```python
df=sbn.load_dataset("titanic")
```

```python
df.head()
```

```python
df.shape
```

# A. Distribution Plots

## 1.Dist-plot

```python
sbn.distplot(df["age"])
```

```python
#For fare
sbn.distplot(df["fare"])
```

```python
#Without KDE
sbn.distplot(df["fare"],kde=False)
```

```python
sbn.displot(df["fare"])
```

```python
sbn.displot(df["age"])
```

## 2.Joint Plot

```python
sbn.jointplot(x=df["age"],y=df["fare"])
```

```python
sbn.jointplot(x=df["age"],y=df["fare"],kind='hex')
```

## 3.Rug plot

```python
sbn.rugplot(data=df,x=df["age"],y=df['fare'],height=0.05)
sbn.scatterplot(data=df, x="age", y="fare")
```

## 4.Pair Plot

```python
sbn.pairplot(df)
```

# <br>

# B.Categorical Data

## 1.Bar-plot

```python
sbn.barplot(data=df,x="sex",y="fare")
```

```python
sbn.barplot(data=df,x="sex",y="fare",estimator=np.std)
```

## B.Count-plot

```python
sbn.countplot(data=df,x="sex")
```

## C.Boxplot

```python
sbn.boxplot(data=df,x="sex",y='age')
```

## D.Violin Plot

```python
sbn.set(style = 'whitegrid') 
sbn.violinplot(data=df,x='sex',y='fare')
```

# C.Advanced Plots

## 1.Strip-plot

```python
sbn.stripplot(data=df,x="sex",y='age')
```

## 2.Sawrmplot

```python
sbn.swarmplot(data=df,x="sex",y="fare",size=5)
```
